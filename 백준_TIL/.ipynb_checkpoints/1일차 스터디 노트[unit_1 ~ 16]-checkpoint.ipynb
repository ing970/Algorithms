```python
w = 10.5
h = 5.5
print(f"width : {w}\nheight : {h}\ntriangle : {w * h / 2}\nSquare : {w * h}")
print("="*40)
print(f"width: {w}, height: {h}\nTriangle : {w * h / 2}, Square: {w * h}")
print("="*40)
print(f"width : {w},\t\t height : {h}\nTriangle : {w * h / 2},\t Sqaure : {w * h}")
```

# 15


```python
user_name = "이석민"
user_age = 27
print("User name : {}".format(user_name))
print("User age : {}".format(user_age))
print("User name: {}, User age: {}".format(user_name, user_age))
```

```python
print("User name : %s" % user_name)
print("User age : %d" % user_age)
print("User name: %s, User age: %d" % (user_name, user_age))

print("Pi : %f" % 3.14)
print("Pi : %d" % 3.14)
```

```python
print("Pi : %.0f" % 3.141592)
print("Pi : %.2f" % 3.141592)
print("Pi : %.4f" % 3.141592)
print("Pi : %.6f" % 3.141592)
```

```python
r = 3
pi = 3.141592
print("반지름 입력 : %d" % r, "\n원주율 입력 : %f" % pi)
print("radius : %.1f" % r, "\npi : %f" % pi)
print("radius : %.1f, pi : %f" % (r, pi))
print("radius : %.6f, pi : %f" % (r, pi))
print("radius : %.2f, pi : %.2f" % (r, pi))
```
